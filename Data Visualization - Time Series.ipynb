# Time Series
- Trends over time
- Tests a hypothesis on a variety of conditions
- Reduces misinterpretation of data

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("countries.csv")
data.head()
```

```python
us = data[data.country == "United States"]
us.head()
```

```python
plt.plot(us.year, us.gdpPerCapita)
plt.title("GDP per capita in the U.S.")
plt.xlabel("year")
plt.ylabel("GDP Per Capita")
plt.show()
```

```python
china = data[data.country == "China"]
china.head()
```

```python
plt.plot(us.year, us.gdpPerCapita)
plt.plot(china.year, china.gdpPerCapita)
plt.legend(['United States', 'China'])
plt.show()
```

```python
us.gdpPerCapita.iloc[0]
```

```python
us_growth = us.gdpPerCapita / us.gdpPerCapita.iloc[0] * 100
china_growth = china.gdpPerCapita / china.gdpPerCapita.iloc[0] * 100
```

```python
plt.plot(us.year, us_growth)
plt.plot(china.year, china_growth)
plt.title("GDP Growth")
plt.ylabel("GDP per capita growth")
plt.xlabel("Year")
plt.show()
```

```python
plt.plot(us.year, us.population)
plt.plot(china.year, china.population)
plt.title("Population growth over the years")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
us_change = us.population / us.population.iloc[0] * 100
china_change = china.population / china.population.iloc[0] * 100

us_change
china_change
```

```python
plt.plot(us.year, us_change)
plt.plot(china.year, china_change)
plt.title("Relative population growth over the years")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```
